# Hangman Env Demo
Quick interactive demo for the Hangman environment.

```python
from pathlib import Path
from utils.helpers import load_corpus, filter_words_by_charset
from env.hangman_env import HangmanEnv
root = Path(__file__).resolve().parents[1] if '__file__' in globals() else Path.cwd()
words = filter_words_by_charset(load_corpus(root / 'data' / 'corpus.txt'))
env = HangmanEnv(words)
obs = env.reset()
print(env.render())
# Make a few random moves
for _ in range(5):
    step = env.step(env.alphabet[_])
    print(step.observation, step.reward, step.done)
    if step.done: break
print('Target:', step.info.get('target'))
```
